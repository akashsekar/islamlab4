```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a DataFrame
data = pd.read_csv('penguins.csv')

# Create a scatterplot for all species
plt.figure(figsize=(10, 6))

# Scatterplot for Adelie penguins
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data[data['species'] == 'Adelie'], label='Adelie', alpha=0.7)

# Scatterplot for Chinstrap penguins
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data[data['species'] == 'Chinstrap'], label='Chinstrap', alpha=0.7)

# Scatterplot for Gentoo penguins
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=data[data['species'] == 'Gentoo'], label='Gentoo', alpha=0.7)

# Add labels and title
plt.title('Relationship between Bill Length and Bill Depth for Penguin Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Bill Depth (mm)')

# Show legend
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a DataFrame
data = pd.read_csv('penguins.csv')

# Calculate BMI for each penguin (BMI = weight_in_g / (height_in_mm^2))
data['BMI'] = (data['body_mass_g'] / ((data['flipper_length_mm'] / 1000) ** 2))

# Group data by gender (male and female) and calculate mean BMI for each group
mean_bmi_by_gender = data.groupby('sex')['BMI'].mean()

# Create a bar chart to visualize mean BMI by gender
plt.figure(figsize=(8, 6))
mean_bmi_by_gender.plot(kind='bar', color=['blue', 'pink'], alpha=0.7)
plt.title('Mean BMI by Gender of Penguins')
plt.xlabel('Gender')
plt.ylabel('Mean BMI')
plt.xticks(rotation=0)  # Ensure gender labels are not rotated
plt.show()
```
